```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, classification_report
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import seaborn as sns
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt'
column_names = ["Variance", "Skewness", "curtosis", "Entropy", "Class"]

df = pd.read_csv(url, names=column_names)
df.head()
sns.pairplot(df, hue="Class")  # label: sınıf (0/1)
plt.show()
```

```python
X = df
y = df["Class"]
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)
```

```python
max_depth = [1, 50, 100, 150]
min_samples_split = [25, 500, 700, 90]
criteria = ['gini', 'entropy']

best_score = 0
best_params = {}

for depth in max_depth:
    for split in min_samples_split:
        for crit in criteria:
            clf = DecisionTreeClassifier(max_depth=depth, min_samples_split=split, criterion=crit)
            clf.fit(X_train, y_train)
            y_pred = clf.predict(X_test)
            acc = accuracy_score(y_test, y_pred)
           # print(f'\nmax_depth={depth}, min_samples_split={split}, criterion={crit}')
            print(classification_report(y_test, y_pred))
            if acc > best_score:
                best_score = acc
                best_params = {
                    'max_depth': depth,
                    'min_samples_split': split,
                    'criterion': crit
                }
print('\nbest results:')
print(f'Acc: {best_score}')
print(f'best: {best_params}')
```

```python
cm = confusion_matrix(y_test, y_pred)

# Görselleştir
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')  # isteğe bağlı renk skalası
plt.title("Confusion Matrix")
plt.show()
```

```python
clf1 = DecisionTreeClassifier(max_depth=3, min_samples_split=5, criterion='gini')
clf1.fit(X_train, y_train)

feature_names = ['variance', 'skewness', 'curtosis', 'entropy', 'class']

plt.figure(figsize=(20,10))
plot_tree(clf1, 
          feature_names=feature_names,  
          class_names=['0', '1'], 
          filled=True)
plt.title("Decision Tree")
plt.show()
```

```python
imp = clf1.feature_importances_
plt.figure(figsize=(10,6))
plt.barh(feature_names, imp, color='skyblue')
plt.xlabel('Feature Importance')
plt.title('Feature Importance Visualization')
plt.show()
```
